```python
from sage.misc.misc_c import prod
from sage.misc.temporary_file import tmp_filename
# from sage.misc.html import html
# from sage.repl.rich_output import pretty_print

from IPython.display import Image
from IPython.core.display import display
# from IPython.core.display import display, HTML

from Hypermatrix_Algebra_tst import HM, Monomial2DiGraph
```

```python
sz = 3; A = HM(sz, sz, 'a')
A.printHM()
```

```python
transMonoid = prod(sum(A[i,j] for j in range(sz)) for i in range(sz)).expand()
```

```python
def my_plot(graph, extra=''):
    tf = tmp_filename(name='fun_graph_', ext=".png")
    graph.plot().save(tf)
    # display(HTML('<img src="http://localhost:8888/tree/tmp_sage_cells/%s" %s>'%(tf.split(r".sage/temp/")[1],extra)))
    display(Image(filename=tf))
    
for graph in transMonoid.operands():
	print(graph)
	my_plot(Monomial2DiGraph(graph, A.list(), sz))
```
