```python
def grouper(iterable, n, fillvalue=None):
    "Collect data into fixed-length chunks or blocks"
    # grouper('ABCDEFG', 3, 'x') --> ABC DEF Gxx"
    args = [iter(iterable)] * n
    return zip(*args)

def my_print(graph, sz, symbol):
    print(graph)
    my_plot(Monomial2DiGraph(graph, HM(sz, sz, symbol).list(), sz))

# Below will still fail because of exponents in L[0] !
# Monomial2DiGraph(L[0], HM(3, 3, 'a_h').list(), 3).plot()

sz = 3;

graphs_on_3 = [[prod(f) for f in grouper(conjclass.arguments(),sz)]  for conjclass in L3]


my_print(graphs_on_3[0][0], sz, 'a_h')

my_print(graphs_on_3[0][1], sz, 'a_l')

my_print(graphs_on_3[1][0], sz, 'a_b')

len(L3)
```
